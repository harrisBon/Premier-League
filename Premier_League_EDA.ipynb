```python
def PointsHome(value):
    if value=="H":
        return 3
    if value== "D":
        return 1
    else :
        return 0
 
df['HP'] = df['FTR'].map(PointsHome)
display(df.head())
```

```python
def PointsAway(value):
    if value=="H":
        return 0
    if value== "D":
        return 1
    else :
        return 3
 
df['AP'] = df['FTR'].map(PointsAway)
display(df.head())
```

```python
b=sns.countplot(x="FTR",data=df)
```

We can see that the number of games that ended with a home win is signifigantly higher than the number of games with the other two results, from that we can get that playing at Home is an important advantage for a team and that it plays a notable role in the outcome of a football game in the English Premier League.

```python
c=sns.countplot(x='FTR',data=df,hue='HTR')
c.set_xlabel("Full Time Result")
c.set_ylabel("Number of Games")
c.set_title("Full Time Result based on Half Time Result")
```

From the above plot we get a couple of insights regarding the affect the Half Time Result has on the end result:
1. We see that if the result at Half-Time is H (meaning the Home Team is winning) then there is a high chance that the match will end with a Home Team win.
2. We see that if the Half-Time result is a Draw then the possiblity for the match ending with any result is almost the same (with a slight advantage again on the Home Team getting a Draw or a Win). For this observation, we can say that it makes sense, since if the result at Half-Time is a draw, then that most likely means that the two teams are equivalent in play at that specific game and therefore the Second-Half can go either way. Of course, there are other aspects that can make two teams equal at Half-Time despite their play, such as potential injuries, yellow or red cards.
3. We can again see that if the Half-Time result is a A (meaning the Away Team is winning) then there is a high chance that the end reult will be the same.

From the above, we can derive that the result at Half-Time is a good indicator of what will be the end result, except when tha first half ends in a draw.

```python
df["SHHTG"]=df["FTHG"]-df["HTHG"]
df["SHATG"]=df["FTAG"]-df["HTAG"]

df.head()
```

```python
print("Half Time Home Team Goals:"+ " " + str(df["HTHG"].sum()))
print("Second Half Home Team Goals:"+ " " + str(df["SHHTG"].sum()))
print("Half Time Away Team Goals:"+ " " + str(df["HTAG"].sum()))
print("Second Half Away Team Goals:"+ " " + str(df["SHATG"].sum()))
```

From the above numbers we get that there are more goals scored during the second half of a game. This occures for various reasons:

1. As the game is going on, the players are getting more tired which may result in a lapse of discipline and focus.

2. Coaches can make substitutions which usually occur during the second half of the game  and also, give more specific tips to their players regarding the game (specifically, during the half-time break).
